Identify missing values and decide on strategies for handling them.

```python
test_df.isnull() #From the output, we can see that the function return dataframe of Boolean values which are True for NaN values.
```

```python
bool_series = pd.isnull(test_df["Cabin"]) #creating bool series True for NaN values
test_df[bool_series] # filtering data and displaying data only with Cabin = NaN
```

Handling of missing values.

```python

test_df['Cabin'].fillna('Staff Cabin', inplace = True) # filling a null values using fillna()
test_df
```

```python
test_df.replace('No Cabin', 'Staff Cabin', inplace=True) #Changing values.
test_df
```

Compute basic statistics (e.g., mean, median, min, max) for numerical columns.

```python
min_value=test_df['Age'].min() # Calculate and print min value of passengers age
print('Minimum Value: '+str(round(min_value, 1)))
```

```python
max_value=test_df['Age'].max() # Calculate and print max value of passengers age
print('Maximum Value: '+str(round(max_value, 1)))
```

```python
mean_value=test_df['Age'].mean() # Calculate and print mean value of passengers age
print('Mean Value: '+str(round(mean_value, 1)))
```

```python
mode_value=test_df['Pclass'].mode() # Calculate and print mode value for passengers class
print('Mode Value: '+str(mode_value))
```

```python
#How many woman on the board
count_value_woman = (test_df['Sex'] == "female").sum()
print('Woman on the board: ' + str(count_value_woman))
```

```python
#How many man on the board
```

```python
count_value_man = (test_df['Sex'] == "male").sum()
print('Man on the board: ' + str(count_value_man))
```

```python
#How many children  under 12 years on the board
count_value_children = (test_df['Age'] < 12).sum()
print('Children age under 12 years on the board: ' + str(count_value_children))
```

```python
#how many age over 62
```

```python
count_value_pensioner = (test_df['Age'] > 62).sum()
print('Passangers age over 62 years on the board: ' + str(count_value_pensioner))
```

Analyze categorical variables (e.g., value counts).

```python
#How many man/woman in each class
sex_pclass_crosstab = pd.crosstab(test_df['Sex'], test_df['Pclass'])
print('Sex and Pclass Crosstab:')
print(sex_pclass_crosstab)
print()
```

Data Visualization:

```python
Filter_children_Frame = test_df[test_df['Age'] < 12]
print(len(Filter_children_Frame))
```

```python
test_df.get(Filter_children_Frame)
print(Filter_children_Frame)
```

```python
plt.bar(Filter_children_Frame['Name'], Filter_children_Frame['Age'], color=['thistle','skyblue'], edgecolor='maroon')
plt.xticks(rotation=90)
plt.title('Children under 12 on the board')
plt.xlabel('Name')
plt.ylabel('Age')
```
